# C. Even, Odd, Positive and Negative


```python
n = int(input())
numbers = list(map(int, input().split()))
even_count = 0
odd_count = 0
positive_count = 0
negative_count = 0
for number in numbers:
    if number % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
    
    if number > 0:
        positive_count += 1
    elif number < 0:
        negative_count += 1

print(f"Even: {even_count}")
print(f"Odd: {odd_count}")
print(f"Positive: {positive_count}")
print(f"Negative: {negative_count}")
```
