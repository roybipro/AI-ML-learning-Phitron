## Keyword Arguments

```python
def adder(a,b,c):
    print(f"{a},{b},{c}")
    
adder(a=3,c=5,b=1)   
```

```python
def adder(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}: {value}")
    
adder(a=12, b=5,c=40, n_estimatoe=50)
```

## Iterator

```python
s ={10,20,30,50}

s_iter = iter(s)


print(next(s_iter))
print(next(s_iter))
print(next(s_iter))
print(next(s_iter))

for value in s:
    print(value)
```

## Generator

```python
list = [x for x in range(500)]

def data_loader(chunk_size,list):
    for i in range(0,len(list),chunk_size):
        yield list[i:i+chunk_size]
        
x = data_loader(5,list)

# print(type(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
```

## Lambda ,Filter ,Apply Function

```python
import pandas as pd

df = pd.DataFrame({'Temperature(F)' : [21,11,89.6,98]})
df

df['Temperature(C)'] = df['Temperature(F)'].apply(lambda x: (x-32) * 5/9) #apply function can use as map

df
```

```python
import pandas as pd
df = pd.DataFrame({'A' : [1,2,3,4,5],'B' : [6,7,8,9,10]})

df['A'] = df['A'].map(lambda x : x ** 0.5)

df
```

```python
import pandas as pd 
df = pd.DataFrame({'Total_spent' : [120,690,300,600]})

#creating a high value feature
df['Is_High_value'] = df['Total_spent'].apply(lambda x : 1 if x>500 else 0)
 
df
```
