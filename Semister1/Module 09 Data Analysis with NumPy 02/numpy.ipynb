## Ndarray Arithmetic Operators and Mathematical function

```python
#faster for vectorized operations
import numpy as np

x = np.array([8, 20, 9, 40, 50])
y = np.array([1, 2, 3, 4, 5])

addition = x + y
print("Addition:", addition)

subtraction = x - y
print("Subtraction:", subtraction)

multiplication = x * y
print("Multiplication:", multiplication)

division = x / y
print("Division:", division)

remainder = x % y
print("Remainder:", remainder)
```

```python
# mathemitical functions

# trigonometric functions

#sin cos tan take radian values as input
sin_values = np.sin(x)
print("Sine values:", sin_values)

cos_values = np.cos(x)
print("Cosine values:", cos_values)

tan_values = np.tan(x)
print("Tangent values:", tan_values)

deg_conversion = np.rad2deg(x) #there are also deg2rad function to convert degree to radian
print("Degree to Radian conversion:", deg_conversion)
```

```python
#log

base_10_log = np.log10(x)
print("Base 10 Logarithm:", base_10_log)

two_log = np.log2(x) #2 diye koto bar vag kora jabe
print("Base 2 Logarithm:", two_log)
```

```python
# square root
sqrt_values = np.sqrt(x)
print("Square Root:", sqrt_values)
```

```python
sum = np.sum(x)
print("Sum of elements in x:", sum)

cumalitive_sum = np.cumsum(x)
print("Cumulative Sum:", cumalitive_sum)
```

### Numpy Logical Functions

```python
#comparison
x = np.array([8, 20, 9, 40, 50])
y = np.array([1, 2, 3, 40, 5])

greater_than = x > y
print("Comparison (x > y):", greater_than)

equal = x == y
print("Comparison (x == y):", equal)    
# all shows if all elements are true
print("All elements in x are greater than y:", np.all(greater_than))
# any shows if any element is true
print("Any element in x is greater than y:", np.any(greater_than))
```

### ndarray Sorting


```python
# inplace sorting
arr = np.array([5, 2, 9, 1, 5, 6])
copy = arr.copy() # this will create a copy of the array
copy.sort() # this will sort the array in place
print("Sorted array:", copy)
print("Original array:", arr) # original array remains unchanged
```

```python
# copy sorting
print("Original array:", arr)
sorted_arr = np.sort(arr) # this will return a sorted copy of the array
print("Sorted array:", sorted_arr)
print("Original array:", arr) # original array remains unchanged
```

```python
#2D array sorting
arr_2d = np.array([[3, 2, 1],
                   [6, 5, 4]])
hor_sorted = np.sort(arr_2d, axis=1) # sort each row
print("Sorted 2D array (row-wise):\n", hor_sorted)
ver_sorted = np.sort(arr_2d, axis=0) # sort each column
print("Sorted 2D array (column-wise):\n", ver_sorted)
```

### ndarray Searching

```python
# np.where
x = np.array([5, 2, 9, 1, 15, 6])

index = np.where(x == 10)
print("Indices of elements equal to 10:", index)

index = np.where(x > 10)
print("Indices of elements greater than 10:", index)

index = np.where(x % 2 == 0)
print("Indices of even elements in x:", index)
index = np.where(x < 8, x, 0) # this will return the elements of x that are less than 8, and 0 for the rest
print("Elements of x less than 8 (else 0):", index)
```

```python
mat = np.array([[10, 20, 30],
                [40, 50, 60]
         ])

index = np.where(mat > 25)
print(index)
```

```python
maximum_values = np.argmax(mat) # this will return the indices of the maximum values along the specified axis
print("Indices of maximum values along columns:", maximum_values)

minimum_values = np.argmin(mat) # this will return the indices of the minimum values along the specified axis
print("Indices of minimum values along columns:", minimum_values)  
```
