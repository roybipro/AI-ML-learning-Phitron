```python
import numpy as np

arr = np.random.randint(1,100,size=(10,5))
```

```python
arr.shape
```

```python
b = arr.reshape(5,10)

b.shape
print(b.ndim)
b
```

```python
flatten = b.flatten()
print(flatten.ndim)

flatten

column_wise_flattening = np.ravel(b, order='C')
print(column_wise_flattening.ndim)
column_wise_flattening
```

```python
a = np.random.randint(1,10,size=(2,3))
b = np.random.randint(20,30,size=(2,3))

print(a)
print(b)   
```

### concatenate

```python
con_row = np.concatenate((a,b), axis=0)  # row wise concatenation
print(con_row)
con_col = np.concatenate((a,b), axis=1)  # column wise concatenation
print(con_col)
```
