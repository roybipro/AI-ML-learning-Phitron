## Ndarray Arithmetic Operators and Mathematical function

```python
#faster for vectorized operations
import numpy as np

x = np.array([8, 20, 9, 40, 50])
y = np.array([1, 2, 3, 4, 5])

addition = x + y
print("Addition:", addition)

subtraction = x - y
print("Subtraction:", subtraction)

multiplication = x * y
print("Multiplication:", multiplication)

division = x / y
print("Division:", division)

remainder = x % y
print("Remainder:", remainder)
```

```python
# mathemitical functions

# trigonometric functions

#sin cos tan take radian values as input
sin_values = np.sin(x)
print("Sine values:", sin_values)

cos_values = np.cos(x)
print("Cosine values:", cos_values)

tan_values = np.tan(x)
print("Tangent values:", tan_values)

deg_conversion = np.rad2deg(x) #there are also deg2rad function to convert degree to radian
print("Degree to Radian conversion:", deg_conversion)
```

```python
#log

base_10_log = np.log10(x)
print("Base 10 Logarithm:", base_10_log)

two_log = np.log2(x) #2 diye koto bar vag kora jabe
print("Base 2 Logarithm:", two_log)
```

```python
# square root
sqrt_values = np.sqrt(x)
print("Square Root:", sqrt_values)
```

```python
sum = np.sum(x)
print("Sum of elements in x:", sum)

cumalitive_sum = np.cumsum(x)
print("Cumulative Sum:", cumalitive_sum)
```
