### Filtering Based on String

```python
import pandas as pd

data ={
    "Name": ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan', 'Judy',],
    "city": ['New York', 'Los Angeles', 'New Delhi', 'Houston', 'Newark', 'New Orleans', 'San Antonio', 'San Diego', 'Dallas', 'H Los Angeles'],
    "Department": ['HR', 'Finance', 'IT', 'Marketing', 'Sales', 'HR', 'Finance', 'IT', 'Marketing', 'Sales'],
    "Salary": [70000, 80000, 90000, 75000, 85000, 72000, 82000, 92000, 77000, 87000]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.loc[df['city'].str.contains("New")]
```

```python
df.loc[df['city'].str.contains("new", case=False)] # case=False means it will ignore the case sensitivity. So it will match "New", "new", "NEW", etc.
```

```python
#start with "Los"
df.loc[df['city'].str.contains(r"^Los")] # ^ means starts with
```

```python
#end with rk
df.loc[df['city'].str.contains(r"rk$")] # $ means ends with 
```

```python
#name start with vowel
df.loc[df['Name'].str.contains(r"^[AEIOUaeiou]")] # ^ means starts with and [AEIOUaeiou] means any of the characters in the brackets
```

```python
#start with New or los
df.loc[df['city'].str.contains(r"(^New|Los)")] # ^ means starts with and (New|Los) means either New or Los
```
