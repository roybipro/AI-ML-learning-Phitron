### 13.1 2D Line Plot


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours = [2 , 3 , 4 , 1 , 5 , 7 , 3]

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.plot(days, hours)
```

```python
plt.xlabel('Days of the week')
plt.ylabel('Hours spent on coding')
plt.title('Trend of coding hours over a week')
plt.plot(days, hours)
```

```python
hours_1st_student = [2 , 3 , 4 , 1 , 5 , 7 , 3]
hours_2nd_student = [3 , 3 , 4 , 2 , 6 , 3 , 2]
hours_3rd_student = [3 , 2 , 5 , 1 , 7 , 6 , 3]

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.xlabel('Days of the week')
plt.ylabel('Hours spent on coding')
plt.title('Trend of coding hours over a week')

plt.plot(days, hours_1st_student, label='1st Student')
plt.plot(days, hours_2nd_student, label='2nd Student')
plt.plot(days, hours_3rd_student, label='3rd Student')
plt.legend()    
```

### 13.3 2D Line from Pandas DF

```python
df = pd.read_csv('enrollment_data.csv')
df.head()
```

```python
plt.xlabel('Year')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrolment over the year')
plt.plot(df['Year'],df['Programming'],label = 'Programming')
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing')
plt.legend()
```

### 13.4 Customizing Line plot ( color,linestyle,marker) and Grid

```python
# 
plt.xlabel('Year')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrolment over the year')
plt.plot(df['Year'],df['Programming'],label = 'Programming',color = 'red',linewidth = 1,linestyle = 'dashdot',marker = 'o',markersize = 9)
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing',color = 'blue',linewidth = 3,linestyle = 'dashed' )
plt.legend()
plt.grid()

plt.show() 
```

### 13.7 Histogram


```python
students = pd.read_csv('student_IQdata.csv')
students.head()
```

```python
# Based on iq
plt.xlabel('IQ Score')
plt.ylabel('Frequency of Students')
plt.title('Distribution of IQ Scores among students')
plt.hist(students['IQ_Score'],color = 'green',edgecolor = 'black')
```

```python
# Based on study hours
plt.xlabel('Study Hours')
plt.ylabel('Frequency of Students')
plt.title('Distribution of Study Hours among students')
plt.hist(students['Study_Hour'],color = 'orange',edgecolor = 'black',bins = 15)
```
