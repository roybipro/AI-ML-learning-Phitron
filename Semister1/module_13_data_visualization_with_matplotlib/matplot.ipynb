### 13.1 2D Line Plot


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours = [2 , 3 , 4 , 1 , 5 , 7 , 3]

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.plot(days, hours)
```

```python
plt.xlabel('Days of the week')
plt.ylabel('Hours spent on coding')
plt.title('Trend of coding hours over a week')
plt.plot(days, hours)
```

```python
hours_1st_student = [2 , 3 , 4 , 1 , 5 , 7 , 3]
hours_2nd_student = [3 , 3 , 4 , 2 , 6 , 3 , 2]
hours_3rd_student = [3 , 2 , 5 , 1 , 7 , 6 , 3]

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.xlabel('Days of the week')
plt.ylabel('Hours spent on coding')
plt.title('Trend of coding hours over a week')

plt.plot(days, hours_1st_student, label='1st Student')
plt.plot(days, hours_2nd_student, label='2nd Student')
plt.plot(days, hours_3rd_student, label='3rd Student')
plt.legend()    
```

### 13.3 2D Line from Pandas DF

```python
df = pd.read_csv('enrollment_data.csv')
df.head()
```

```python
plt.xlabel('Year')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrolment over the year')
plt.plot(df['Year'],df['Programming'],label = 'Programming')
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing')
plt.legend()
```

### 13.4 Customizing Line plot ( color,linestyle,marker) and Grid

```python
# 
plt.xlabel('Year')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrolment over the year')
plt.plot(df['Year'],df['Programming'],label = 'Programming',color = 'red',linewidth = 1,linestyle = 'dashdot',marker = 'o',markersize = 9)
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing',color = 'blue',linewidth = 3,linestyle = 'dashed' )
plt.legend()
plt.grid()

plt.show() 
```

### 13.8 Bar Chart


```python
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan', 'Judy'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female'],
    'Study_Hour': [2, 3, 4, 1, 5, 7, 3, 6, 2, 4],
    'Attendance': [90, 85, 95, 80, 92, 88, 91, 87, 89, 55],
    'Grade': ['Pass', 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Pass', 'Fail', 'Pass']
}

df = pd.DataFrame(data)
df.head()
```

```python
# gender wise group
gender_group = df.groupby('Gender').size()

gender_group
```

```python
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar chart of Gender among students')
plt.bar(gender_group.index,gender_group.values,color=['orange','red'],edgecolor = 'black')
```

```python
#bar Chart on real data

df = pd.read_csv('student_data.csv')
df.head()
```

```python
status = df.groupby('CompletionStatus').size()

plt.xlabel('Completion Status')
plt.ylabel('Count')
plt.title('Bar Chart on Completion status')
plt.bar(status.index,status.values, color = ['green','orange','red'], edgecolor = 'black')
plt.show()
```

```python
location = df.groupby('Location').size()
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Bar Chart on Location')
plt.bar(location.index,location.values, color = ['green','orange','red','pink','blue'], edgecolor = 'black')
plt.show()
```

```python
Instructor = df.groupby('Instructor').size()
plt.xlabel('Instructor')
plt.ylabel('Count')
plt.title('Bar Chart on Instructor')
plt.bar(Instructor.index,Instructor.values, color = ['green','orange','red'], edgecolor = 'black')
plt.show()
```

### 13.9 Pie Chart

```python
status = df.groupby('CompletionStatus').size()
plt.pie(status,labels=status.index,autopct='%1.1f%%',colors = ['green','orange','red'],explode = [0.1,0.1,0.1],shadow = True)
plt.show()
```

```python
location = df.groupby('Location').size()
plt.pie(location,labels=location.index,autopct='%1.1f%%',explode = [0.1,0,0,0,0],colors = ['green','orange','red','blue','purple'],shadow = True)
plt.show()
```
