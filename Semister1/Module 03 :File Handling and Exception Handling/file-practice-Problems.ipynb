## Practice 1 :

Get the text from sample.txt , and count the following


Number of lines

Number of words

Number of characters

and save it to the counter_of_string.txt file

```python
from functools import reduce
with open("../sample_data/test.txt","r") as file:
    
    #number of lines
    string_list = file.readlines()
    total_lines = len(string_list)
    print("Total Lines :",total_lines)
    
    #Number of words
    number_of_words = list(map(lambda x : len(x.split()),string_list))
    print("Number Of words",number_of_words)
    
    #total number of words
    total_numbers_words = reduce(lambda x,y : x+y ,number_of_words)
    print("Total Number Of Words :",total_numbers_words)
    
    #total number of characters
    
    #cleaning process
    
    # 1.delete new line
    string_list = list(map(str.strip,string_list))
    #2.space delete
    string_list = map(lambda x :x.replace(" ",""),string_list)
    print("Number of Character",string_list)
    
    number_of_character = list(map(lambda x : len(x),string_list))
    print("Number of Character",number_of_character)

    #total number of characters
    total_numbers_character = reduce(lambda x,y : x+y ,number_of_character)
    print("total number of characters :",total_numbers_character)

#save it
with open("../sample_data/counter_of_string.txt","w") as file:
    file.write(f"total line :{total_lines}\n total number of words :{total_numbers_words}\n total number of characters :{total_numbers_character}")

    
     
```
