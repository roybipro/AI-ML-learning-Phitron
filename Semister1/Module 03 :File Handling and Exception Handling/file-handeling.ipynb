## Read file

```python
file = open("../sample_data/test.txt", "r")

content = file.read()

print(content)
```

```python
file = open("../sample_data/test.txt", "r")

content = file.readlines()

print(content)

file.close()
print(file.closed)
```

```python
with open("../sample_data/test.txt", "r") as file:
  for line in file:
      print(line,end="")

print(file.closed)
```

## File pointer

```python
with open("../sample_data/test.txt", "r") as file:
    print(file.tell()) #currently cursor
    
    print(file.read(5))
    
    print(file.tell())
    
    print(file.read())
```

```python
#Seek
with open("../sample_data/test.txt", "r") as file:
    print(file.tell())
    file.seek(5)
    print(file.tell())
    
    file.seek(0)
    print(file.tell())
    print(file.read(5))
```
