```python
# Ndarray

import numpy as np

# 1d array

arr1 = np.array([1,2,3,4,5])

print(type(arr1))

# 2d array
arr2 = np.array([[1,2,3], #row 1
                 [4,5,6]]) #row 2


#3d array
arr3 = np.array([
                [ [1,2,3],
                  [4,5,6] ],
                
                [ [1,2,3],
                  [4,5,6] ]
                 
                 ])
```

```python
#array attributes

# dimentions
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

#shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

#data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

#size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

## Ndarray Data type ##

```python
import numpy as np

arr = np.array([1,2,3])
print(arr.dtype)

arr = np.array([1.3,3.5,3.4])
print(arr.dtype)

arr = np.array([1,2,3.3])
print(arr.dtype) #upcasting 

arr = np.array([1,2,'B'])
print(arr.dtype) #upcasting to string
```

selecting a data type for an array

```python
arr = np.array([1,2,4])
print(arr.dtype)
arr = arr.astype(np.int16)
print(arr.dtype)
```

## Ndarray creation from existing data

```python
#list
import numpy as np
lst = [10,20,30,40,50.6]

arr = np.array(lst, dtype=np.int32)

print(arr)
print(type(arr))
print(arr.dtype)

mixed_lst = [10,True,'hello', 3.14 ]

arr = np.array(mixed_lst)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)

matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]
arr = np.array(matrix)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)

```

```python
#Tuples 
import numpy as np
tup = (10,20,30,40,50.6)

arr = np.array(tup)  
arr = arr.astype(np.int16) 

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
#Dictionaries
import numpy as np
dct = {'a':10, 'b':20, 'c':30}

keys = dct.keys()
values = dct.values()
print(keys)
print(values)
items = dct.items()
print(items)

# keys to ndarray
arr = np.array(list(keys))
print(arr)
print(type(arr))
# arr = arr.astype(np.int16)
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```
