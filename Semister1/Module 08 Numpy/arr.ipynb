```python
# Ndarray

import numpy as np

# 1d array

arr1 = np.array([1,2,3,4,5])

print(type(arr1))

# 2d array
arr2 = np.array([[1,2,3], #row 1
                 [4,5,6]]) #row 2


#3d array
arr3 = np.array([
                [ [1,2,3],
                  [4,5,6] ],
                
                [ [1,2,3],
                  [4,5,6] ]
                 
                 ])
```

```python
#array attributes

# dimentions
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

#shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

#data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

#size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```
