```python

class Phone:
    category = "Electronics"

    # constructor
    def __init__(self, model, battery, camera, battery_percentage=100):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage  

    # methods
    def charge(self, hour):
        self.battery_percentage = min(100, self.battery_percentage + hour)  

    def Capture(self, photo):
        if self.battery_percentage <= 0:
            print("No Charge")
        elif self.battery_percentage < photo:
            print("Not enough charge")
        else:
            self.battery_percentage -= photo
            print(f"Photo Captured in {self.model}")
```

```python
## creating object
apple = Phone("Iphone 17 pro",4000,48,100)
blueberry = Phone("B-17",4000,24)
motorola = Phone("M max",3500,40)

print(motorola.camera)

motorola.Capture(10)
print(motorola.battery_percentage)
print(apple.battery_percentage)
```
