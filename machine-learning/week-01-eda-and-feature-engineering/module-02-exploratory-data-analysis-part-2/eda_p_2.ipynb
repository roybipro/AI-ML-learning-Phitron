```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_data.csv')
df.head()
```

### 2.2 Multivariate Analysis- 1 ( Countplot with Hue-1)

```python
sns.countplot(x=df['Sex'],hue=df['Survived'])
```

```python
gender_split = df.groupby('Sex')['Survived'].value_counts(normalize=True)
print(gender_split)
```

### 2.3 Multivariate Analysis -1 ( Countplot with Hue-2)

```python
sns.countplot(x=df['Pclass'],hue=df['Survived'])
```

```python
pclass_split = df.groupby('Pclass')['Survived'].value_counts(normalize=True)
print(pclass_split)
```
